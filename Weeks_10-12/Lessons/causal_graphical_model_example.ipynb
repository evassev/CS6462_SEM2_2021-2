# Causal Graphical Models in Python

In this example, we create a causal graphical model with Python. The causal network depicts the Sprinkler example covered by the lectures.   

In this example, we use the causalgraphicalmodels Python library. You can find this library here: https://github.com/ijmbarr/causalgraphicalmodels

## Step 1. Create the DAG of the Causal Network

Use the CausalGraphicalModel class to create the DAG of the "Sprinkler" Causal Network:
<li>provide an array of nodes</li>
<li>provide an array of edges</li>

```python
from causalgraphicalmodels import CausalGraphicalModel

sprinkler = CausalGraphicalModel(
    nodes=["season", "rain", "sprinkler", "wet", "slippery"],
    edges=[
        ("season", "rain"), 
        ("season", "sprinkler"), 
        ("rain", "wet"),
        ("sprinkler", "wet"), 
        ("wet", "slippery")
    ]
)
```

## Step 2. Show the DAG of the Causal Network

```python
# draw the graphical model
sprinkler.draw()
```

## Step 3. Show the Joint Dstribution of the Causal Network

```python
print(sprinkler.get_distribution())
```

## Step 4. Run the intervention do("sprinkler") on the Causal Network

```python
sprinkler_do = sprinkler.do("sprinkler")
```

## Step 5. Show the Causal DAG with the intervention "on"

```python
sprinkler_do.draw()
```

## Step 6. Show the Joint Distibution of the Causal Network with the intervention "on"

```python
print(sprinkler_do.get_distribution())
```

## Compare Joint Distributions
Joint distribution without interventions: <br>
<b>P(season)P(rain|season)P(sprinkler|season)P(wet|rain,sprinkler)P(slippery|wet)</b>
<br><br>
Joint distribution with <i>do("sprinkler")</i> intervention: <br>
<b>P(season)P(rain|season)P(wet|rain,do(sprinkler))P(slippery|wet)</b>
